```scala
%spark.pyspark

from pyspark.sql import functions as F

users = spark.createDataFrame(
    [("u1", "Berlin"),
     ("u2", "Berlin"),
     ("u3", "Munich"),
     ("u4", "Hamburg")],
    ["user_id", "city"]
)

orders = spark.createDataFrame(
    [("o1", "u1", "p1", 2, 10.0),
     ("o2", "u1", "p2", 1, 30.0),
     ("o3", "u2", "p1", 1, 10.0),
     ("o4", "u2", "p3", 5, 7.0),
     ("o5", "u3", "p2", 3, 30.0),
     ("o6", "u3", "p3", 1, 7.0),
     ("o7", "u4", "p1", 10, 10.0)],
    ["order_id", "user_id", "product_id", "qty", "price"]
)

products = spark.createDataFrame(
    [("p1", "Ring VOLA"),
     ("p2", "Ring POROG"),
     ("p3", "Ring TISHINA")],
    ["product_id", "product_name"]
)

users.show()
orders.show()
products.show()
```

```scala
%spark.pyspark
from pyspark.sql import functions as F
from pyspark.sql.window import Window


orders_with_rev = orders.withColumn("revenue", F.col("qty") * F.col("price"))


enriched = (
    orders_with_rev
    .join(users, on="user_id", how="inner")
    .join(products, on="product_id", how="inner")
)

enriched.show(truncate=False)
```

```scala
%spark.pyspark

data_mart = (
    enriched
    .groupBy("city", "product_id", "product_name")
    .agg(
        F.countDistinct("order_id").alias("orders_cnt"),
        F.sum("qty").alias("qty_sum"),
        F.sum("revenue").alias("revenue_sum"),
    )
)
```

```scala
%spark.pyspark
w = Window.partitionBy("city").orderBy(F.col("revenue_sum").desc(), F.col("product_id").asc())

mart_top2 = (
    data_mart
    .withColumn("rn", F.row_number().over(w))
    .filter(F.col("rn") <= 2)
    .drop("rn")
)
mart_top2.orderBy("city", F.col("revenue_sum").desc()).show(truncate=False)
```

```scala
%spark.pyspark


hdfs_path = "/tmp/sandbox_zeppelin/mart_city_top_products/"

(
    mart_top2
    .write
    .mode("overwrite")
    .parquet(hdfs_path)
)

print("Saved to HDFS:", hdfs_path)
```

```scala
%spark.pyspark
from pyspark.sql import functions as F

s3_path = "s3a://hadoopyandex/mart_city_top_products/"
hdfs_path = "/tmp/sandbox_zeppelin/mart_city_top_products/"

print("=== Read back from S3 ===")
mart_s3 = spark.read.parquet(s3_path)
mart_s3.orderBy("city", F.col("revenue_sum").desc()).show(truncate=False)

print("=== Read back from HDFS ===")
mart_hdfs = spark.read.parquet(hdfs_path)
mart_hdfs.orderBy("city", F.col("revenue_sum").desc()).show(truncate=False)

print("=== Global top-1 by revenue_sum (from S3) ===")
mart_s3.orderBy(F.col("revenue_sum").desc()).show(1, truncate=False)
```
